```python
import time
import random
from bs4 import BeautifulSoup
import requests
import pandas as pd

title = []
synopsis = []

# Here are the things you will have to do for all links:
# - Slow down the frequency of requests to avoid being identified and therefore banned from the website.
# Use `time.sleep(random.uniform(1.0, 2.0))`
# - Get request object from URL
# - Extract the content into a variable using BeautifulSoup
# - Get title
# - Get synopsis

for link in movie_links:
    # Slow down the frequency of requests
    time.sleep(random.uniform(1.0, 2.0))
    
    # Get request object from URL
    response = requests.get(link)
    
    # Extract the content into a variable using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Get title
    title_tag = soup.find('h1', class_='title')
    title.append(title_tag.text.strip() if title_tag else 'N/A')
    
    # Get synopsis
    synopsis_tag = soup.find('div', class_='synopsis')
    synopsis.append(synopsis_tag.text.strip() if synopsis_tag else 'N/A')

# Check the length of the lists before creating the dataframe
print("Length of title list:", len(title))
print("Length of synopsis list:", len(synopsis))
print("Length of movie_links list:", len(movie_links))

# Create DataFrame
df = pd.DataFrame({"Title": title, "Synopsis": synopsis, "Links": movie_links})

# Save DataFrame to CSV file
df.to_csv("./assets/allo_cine.csv", index=False)
```
